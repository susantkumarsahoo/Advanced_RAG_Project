```python
import boto3
import os
from dotenv import load_dotenv

# Load keys from the .env file
load_dotenv()
```

```python

# Connect to S3
s3 = boto3.client(
    's3',
    aws_access_key_id=os.getenv('AWS_ACCESS_KEY_ID'),
    aws_secret_access_key=os.getenv('AWS_SECRET_ACCESS_KEY'),
    region_name=os.getenv('AWS_REGION')
)

# Test by listing your buckets
try:
    response = s3.list_buckets()
    print("Connection Successful! Your buckets:")
    for bucket in response['Buckets']:
        print(f"- {bucket['Name']}")
except Exception as e:
    print(f"Error: {e}")
```

```python
# 3. Set your bucket and file details
bucket_name = 'rag-s3-docs-bucket'
local_file = r'C:\Users\TPWODL\New folder_Content\Advanced_RAG_Project\data\raw\Supply Code, 2019 with Gazette.pdf'      # The name of the file on your PC
s3_destination = 'data/SupplyCode.pdf'  # The name/path it will have in S3

try:
    # 4. Upload the file
    print(f"Uploading {local_file}...")
    s3.upload_file(local_file, bucket_name, s3_destination)
    print("Upload Successful!")

except FileNotFoundError:
    print("Error: The local PDF file was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```
